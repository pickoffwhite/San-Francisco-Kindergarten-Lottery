Analysing how many people put down more than 10 school choices.

```python
amountchoices = lotto.drop(['StudentNo', "Does Student Live In CTIP1 Zone? (y/n)",'School Enrolled In As Of 11/03/2017', 'Round 1 Assignment', "Student's Ethnicity", "Student's Residential Zip Code", 'got1stchoice', 'gottop5', 'gottop10'], axis=1)
```

```python
lotto['choices'] = amountchoices.notnull().sum(axis=1)
```

```python
lotto['morethan10'] = np.where(lotto['choices'] > 10, 'yes', 'no')
```

```python
lotto.morethan10.value_counts('yes')
```

Examining how many people put down duplicate choices.

```python
duplicates = lotto.drop(['StudentNo', "Does Student Live In CTIP1 Zone? (y/n)",'School Enrolled In As Of 11/03/2017', 'Round 1 Assignment', "Student's Ethnicity", "Student's Residential Zip Code", 'got1stchoice', 'gottop5', 'gottop10', 'choices', 'morethan10' ], axis=1)
```

```python
duplicates['uniquechoices'] = duplicates.apply(lambda x: x.nunique(), axis=1)
```

```python
lotto['amountduplicates'] = lotto['choices'] - duplicates['uniquechoices']
```

```python
lotto['duplicates'] = np.where(lotto['amountduplicates'] > 0, 'yes', 'no')
```

```python
lotto.duplicates.value_counts('no')
```

Does putting down more than ten schools impact the chance of enrolling in your first choice?

```python
lotto.groupby(['got1stchoice', 'morethan10']).size()
```

Does putting down duplicates impact the chance of enrolling in your first choice?

```python
lotto.groupby(['got1stchoice', 'duplicates']).size()
```
